```python
### for Des we take 8 byte key
data = input("Enter Plain text========")
keystring = input("enter 8 byte string for key generation========")
k= pyDes.des(keystring,padmode=pyDes.PAD_PKCS5)
e=k.encrypt(data)
print("cipher text=== %r" %e)
print("plain text=== %r" %k.decrypt(e))
t1=time()
for i in range(1000):
  e=k.encrypt(data)
t2=time()
print("time for 1000 encryption:{:0.3f}s".format(t2-t1))

```

```python
### For Triple DES we are taking 16/32 byte key
data = input("Enter Plain text========")
keystring = input("enter 16/32 byte string for key generation========")
k= pyDes.triple_des(keystring,padmode=pyDes.PAD_PKCS5)
e=k.encrypt(data)
print("cipher text=== %r" %e)
print("plain text=== %r" %k.decrypt(e))
t1=time()
for i in range(1000):
  e=k.encrypt(data)
t2=time()
print("time for 1000 encryption:{:0.3f}s".format(t2-t1))
```
