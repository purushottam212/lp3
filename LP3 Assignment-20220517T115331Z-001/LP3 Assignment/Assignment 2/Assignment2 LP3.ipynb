```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import datetime as dt
%matplotlib inline
dataset=pd.read_csv("Cosmetic shop.csv")
dataset
```

```python
X=dataset.iloc[:,:-1]
X
```

```python
y=dataset.iloc[:,5]
y
```

```python
#perform label encoding to convert categorical data into numerical data
from sklearn.preprocessing import LabelEncoder
labelencoder_x=LabelEncoder()#we have created the object of label encoder
```

```python
X=X.apply(LabelEncoder().fit_transform) # we provided input as 'x' values to the label encoder
print(X)
```

```python
from sklearn.tree import DecisionTreeClassifier
regressor=DecisionTreeClassifier()
regressor.fit(X.iloc[:,1:5],y)
```

```python
#predict value for the given Expression
x_in=np.array([1,1,0,0])
y_pred=regressor.predict([x_in])
print("Prediction:",y_pred)
```
